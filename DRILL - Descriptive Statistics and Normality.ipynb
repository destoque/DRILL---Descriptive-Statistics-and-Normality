```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#lognormal ditribution

lognormal = np.random.lognormal(5, 1, 100)

# Add a vertical line at the mean.
plt.axvline(lognormal.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(lognormal.mean() + lognormal.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(lognormal.mean()-lognormal.std(), color='b', linestyle='dashed', linewidth=2) 

#The descriptive statistics did not provide useful information about the variable as
#the data points wre spread out among the graph

plt.hist(lognormal)
plt.show()
```

```python
#multinomial ditribution

multinomial = np.random.multinomial(100, [1/6]*6, size=1)

# Add a vertical line at the mean.
plt.axvline(multinomial.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(multinomial.mean() + multinomial.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(multinomial.mean()-multinomial.std(), color='b', linestyle='dashed', linewidth=2) 

#The descriptive statistics did not provide useful information about the variable as
#the data points were very spread out among the graph

plt.hist(multinomial)
plt.show()
```

```python
#pareto ditribution

pareto = np.random.pareto(100, 100)

# Add a vertical line at the mean.
plt.axvline(pareto.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(pareto.mean() + pareto.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(pareto.mean()-pareto.std(), color='b', linestyle='dashed', linewidth=2) 

#The descriptive statistics did not provide useful information about the variable as
#the data points were very spread out among the graph. The information is skewed to the left
#The standard deviation is quite large

plt.hist(pareto)
plt.show()
```

```python
#Uniform ditribution

uniform = np.random.uniform(0, 100, 100)

# Add a vertical line at the mean.
plt.axvline(uniform.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(uniform.mean() + uniform.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(uniform.mean()-uniform.std(), color='b', linestyle='dashed', linewidth=2) 

#The descriptive statistics did not provide useful information about the variable as
#the data points all have an equal chance of being drawn. There for the  mean will
#always be 50

plt.hist(uniform)
plt.show()
```

```python
#standard expoential distribution

standard_exponential = np.random.standard_exponential(100)

# Add a vertical line at the mean.
plt.axvline(standard_exponential.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(standard_exponential.mean() + standard_exponential.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(standard_exponential.mean()-standard_exponential.std(), color='b', linestyle='dashed', linewidth=2) 

#The descriptive statistics for this distribution provide an acccurate
#prediction of what the numbers will be if ran again

plt.hist(standard_exponential)
plt.show()
```

```python
#geometric distribution

geometric = np.random.geometric(0.2 ,100)

# Add a vertical line at the mean.
plt.axvline(geometric.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(geometric.mean() + geometric.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(geometric.mean()-geometric.std(), color='b', linestyle='dashed', linewidth=2) 

#The geometric distribution models the number of trials that must be run in order to achieve success.
#meaning the lower the probability of success the more trials need to be run

plt.hist(geometric)
plt.show()
```

```python
#second part of the drill

variable1 = np.random.normal(5,.5, 1000)
variable2 = np.random.normal(10,1, 1000)
variable3 = variable1+variable2

mean = np.mean(variable3)
sd = np.std(variable3)

plt.hist(variable3)
plt.axvline(x=mean,color='b', linestyle='solid')
plt.axvline(x=mean+sd, color='b', linestyle='dashed')
plt.axvline(x=mean-sd, color='b', linestyle='dashed')
plt.show()
```
